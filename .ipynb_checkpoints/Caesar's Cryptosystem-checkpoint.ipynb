# Caesar's Cryptosystem

```python
def crypto(msg):
    x = list(msg)
    alphabet = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
    cryptic = [alphabet[(j+3)%26] for i,vali in enumerate(x) for j,valj in enumerate(alphabet) if vali==valj]
    return "".join(cryptic)
```

```python
def run(msg):
    words = msg.split(" ")
    final = []
    for word in words:
        final.append(crypto(word.lower()))
        final.append(" ")
    return ''.join(final)
```

```python
run("Hello World")
```
